```python
# Import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load gambar
img_path = 'Lenna.png'
img = cv2.imread(img_path)
# Mengubah gambar dari BGR ke RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR ke RGB

# Menampilkan gambar asli
plt.title("Lenna Original Image")
plt.imshow(img_rgb)
plt.axis('off') # untuk menghilangkan sumbu
plt.show()

# Lena grayscale
lenna = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.title("Lenna Grayscale")
plt.imshow(lenna, cmap='gray')
plt.axis('off')
plt.show()
```

# Threshold
Thresholding adalah teknik pemrosesan citra yang digunakan untuk mengubah citra grayscale menjadi citra biner (hitam-putih) dengan membandingkan nilai pixel dengan nilai ambang tertentu (threshold). 

Jika nilai pixel lebih besar dari threshold, pixel tersebut akan menjadi putih (255), sedangkan jika lebih kecil, pixel tersebut akan menjadi hitam (0)

```python
# Langkah-Langkah:
# 1. Convert gambar ke grayscale
lenna = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.title("Lenna Grayscale")
plt.imshow(lenna, cmap='gray')
plt.axis('off')
plt.show()

# 2. Lakukan thresholding
_, threshold_img = cv2.threshold(lenna, 127, 255, cv2.THRESH_BINARY)
plt.title("Lenna Thresholding")
plt.imshow(threshold_img, cmap = 'gray')
plt.axis('off')
plt.show()
```

# Negative
Operasi negative pada gambar adalah proses membalikkan nilai-nilai pixel gambar sehingga warna yang terang menjadi gelap, dan yang gelap menjadi terang.

```python
# Operasi mengubah gambar menjadi negative
negative_image = 255 - lenna

# Lenna kondisi awal (GrayScale)
plt.title('Origial Image Gray')
plt.imshow(lenna, cmap='gray')
plt.axis('off')
plt.show()

# Menampilkan hasil
plt.title('Negative Image')
plt.imshow(negative_image,  cmap = 'gray')
plt.axis('off')
plt.show()
```

# Kecerahan
Operasi kecerahan (brightness adjustment) adalah proses meningkatkan atau menurunkan intensitas cahaya dari sebuah gambar, yaitu dengan menambahkan atau mengurangkan nilai piksel gambar grayscale atau berwarna.

```python
# Misalnya lenna sudah dalam grayscale
bright_image = cv2.convertScaleAbs(lenna, alpha=1, beta=80)

# Lenna kondisi awal (GrayScale)
plt.title('Origial Image Gray')
plt.imshow(lenna, cmap='gray')
plt.axis('off')
plt.show()

# Menampilkan hasi
plt.imshow(bright_image, cmap = 'gray')
plt.title('Lenna Brightness Up')
plt.axis('off')
plt.show()
```
