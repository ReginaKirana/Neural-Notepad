```python
# Import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load gambar
img_path = 'Lenna.png'
img = cv2.imread(img_path)
# Mengubah gambar dari BGR ke RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR ke RGB

# Menampilkan gambar asli
plt.title("Lenna Original Image")
plt.imshow(img_rgb)
plt.axis('off') # untuk menghilangkan sumbu
plt.show()
```

```python
# Lena grayscale
lenna_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(lenna_gray, cmap='gray')
plt.title("Lenna Grayscale")
plt.axis('off')
plt.show()
```

# Threshold
Thresholding adalah teknik pemrosesan citra yang digunakan untuk mengubah citra grayscale menjadi citra biner (hitam-putih) dengan membandingkan nilai pixel dengan nilai ambang tertentu (threshold). 

Jika nilai pixel lebih besar dari threshold, pixel tersebut akan menjadi putih (255), sedangkan jika lebih kecil, pixel tersebut akan menjadi hitam (0)

```python
def thresholding(x, T):
    x = x.astype(np.float64)  # konversi ke float
    y = np.zeros_like(x)

    baris, kolom = x.shape
    for i in range(baris):
        for j in range(kolom):
            if x[i, j] < T:
                y[i, j] = 0
            else:
                y[i, j] = 255
    return y.astype(np.uint8)

# Lena grayscale
lenna_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(lenna_gray, cmap='gray')
plt.title("Lenna Grayscale (Initial State)")
plt.axis('off')
plt.show()

# Terapkan thresholding dengan T = 125
thresh_result = thresholding(lenna_gray, 125)
# Tampilkan hasil
plt.imshow(thresh_result,  cmap='gray')
plt.title("Lenna Thresholding Result")
plt.axis('off')
plt.show()
```

# Negative
Operasi negative pada gambar adalah proses membalikkan nilai-nilai pixel gambar sehingga warna yang terang menjadi gelap, dan yang gelap menjadi terang.

```python
# Lena grayscale
lenna_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(lenna_gray, cmap='gray')
plt.title("Lenna Grayscale (Initial State)")
plt.axis('off')
plt.show()

def negatif(x):
    x = x.astype(np.float64)  # Konversi ke float (mirip double)
    y = np.zeros_like(x)

    baris, kolom = x.shape
    for i in range(baris):
        for j in range(kolom):
            y[i, j] = 255 - x[i, j]

    return y.astype(np.uint8)

negatif_lenna = negatif(lenna_gray)
plt.imshow(negatif_lenna, cmap = 'gray')
plt.axis('off')
plt.title('Lenna Negatif')
plt.show()
```

# Kecerahan
Operasi kecerahan (brightness adjustment) adalah proses meningkatkan atau menurunkan intensitas cahaya dari sebuah gambar, yaitu dengan menambahkan atau mengurangkan nilai piksel gambar grayscale atau berwarna.

```python
def kecerahan(x, k):
    x = x.astype(np.int16)  # Agar aman dari overflow
    y = np.zeros_like(x)

    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            y[i, j] = min(max(x[i, j] + k, 0), 255)  # Clamp nilai ke 0-255

    return y.astype(np.uint8)

# Lena grayscale
lenna_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(lenna_gray, cmap='gray')
plt.title("Lenna Grayscale (Initial State)")
plt.axis('off')
plt.show()

# Implementasikan Kecerahan
lenna_terang = kecerahan(lenna_gray, 80)
plt.imshow(lenna_terang, cmap = 'gray')
plt.title('Lenna Brightness Up')
plt.axis('off')
plt.show()

lenna_gelap = kecerahan(lenna_gray, -100)
plt.imshow(lenna_gelap, cmap = 'gray')
plt.title('Lenna Brightness Down')
plt.axis('off')
plt.show()
```
